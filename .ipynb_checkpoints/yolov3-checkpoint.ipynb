Helper Functions to 
1. Show Image
2. Download Image

```python
from IPython import display
def imShow(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(10, 6)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()

# use this to download a file  
def download(path):
  from google.colab import files
  files.download(path)
```

```python
# run darknet detection
!./darknet detect cfg/yolov3.cfg yolov3.weights data/person.jpg
display.clear_output(wait = True)
imShow('predictions.jpg')
```

1. After copying yhe data to you VM, now you have to check that the data is in it's right place which is /darknet/data/name_of_your_dir.
2. Then change the name_of_your_dir to obj.

Copy the Yolo Config file to drive or download it to you machine and make some edits so to make it ready for training. (N = number of classes you're training on)
1. Comment Testing, Subdivision, and Batches that follows Testing
2. Uncomment Training, Subdivisions, and Batches that follows Training.
3. Change the number of max_batches to 2000*N
4. Change the steps to 80% and 90% of total Iterations.
5. Search for YOLO in the file and Change the number of classes in that Yolo sections to the number of classes you want to train on. (N = number of classes you're training on)
6. In the Convolutional layer above that change the number of filters to ( N + 5 ) * 3. Do this for all 3 Yolo Sections
